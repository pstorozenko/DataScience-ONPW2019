```python
weather['tempC'] = (weather.temp - 32) * 5/9
```

```python
weather['datetime'] = pd.to_datetime(weather.time_hour)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(weather.datetime, weather.tempC)
```

```python
weatherdt = weather.set_index('datetime')
```

```python
weatherdt_mean = weatherdt.resample('D').mean()
```

```python
plt.plot(weatherdt_mean.index, weatherdt_mean.tempC)
```

```python
temp_max = weatherdt_mean.tempC.nlargest(10).to_frame('tempC')
```

```python
plt.plot(weatherdt_mean.index, weatherdt_mean.tempC)
plt.plot(temp_max.index, temp_max.tempC, ' o')
```
